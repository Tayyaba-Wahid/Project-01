```python
#Inventory Management System
import os

inventory = {}

def add_item():
    name = input("Enter item name: ")
    quantity = int(input("Enter quantity: "))
    price = float(input("Enter price: "))
    category = input("Enter category: ")

    if name in inventory:
        print("Item already exists. Please update the quantity instead.")
    else:
        inventory[name] = {
            "quantity": quantity,
            "price": price,
            "category": category
        }
        print("Item added successfully!")

def search_item():
    search_term = input("Enter item name or category: ")
    found_items = []

    for name, details in inventory.items():
        if search_term.lower() in name.lower() or search_term.lower() in details["category"].lower():
            found_items.append((name, details))

    if found_items:
        for item in found_items:
            print(f"Name: {item[0]}, Quantity: {item[1]['quantity']}, Price: {item[1]['price']}, Category: {item[1]['category']}")
    else:
        print("No items found.")

def update_stock():
    name = input("Enter item name: ")
    if name in inventory:
        quantity = int(input("Enter new quantity: "))
        inventory[name]["quantity"] = quantity
        print("Quantity updated successfully!")
    else:
        print("Item not found.")

def delete_item():
    name = input("Enter item name: ")
    if name in inventory:
        del inventory[name]
        print("Item deleted successfully!")
    else:
        print("Item not found.")

def view_total_value():
    total_value = sum(item["quantity"] * item["price"] for item in inventory.values())
    print(f"Total inventory value: {total_value}")

def save_inventory():
    with open("inventory.txt", "w") as file:
        for name, details in inventory.items():
            file.write(f"{name},{details['quantity']},{details['price']},{details['category']}\n")

def load_inventory():
    global inventory
    if os.path.exists("inventory.txt"):
        with open("inventory.txt", "r") as file:
            for line in file.readlines():
                name, quantity, price, category = line.strip().split(",")
                inventory[name] = {
                    "quantity": int(quantity),
                    "price": float(price),
                    "category": category
                }

def main():
    load_inventory()
    while True:
        print("1. Add item")
        print("2. Search item")
        print("3. Update stock")
        print("4. Delete item")
        print("5. View total inventory value")
        print("6. Save and exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            add_item()
        elif choice == "2":
            search_item()
        elif choice == "3":
            update_stock()
        elif choice == "4":
            delete_item()
        elif choice == "5":
            view_total_value()
        elif choice == "6":
            save_inventory()
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
